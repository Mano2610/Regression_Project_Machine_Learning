Since the dataset has two object(categorical) datatypes, Lets use label encoder to make a dummy variables

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
df['F20'] = le.fit_transform(df['F20'])
df['F27'] = le.fit_transform(df['F27'])
```

Now, all the data types are numerical. sklearn models only accepts the dataset of numerical values.

```python
df.describe()
```

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in[~((df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)).all(axis=1)]
    return df_out
```

```python
#Function to correlate the data with the threshold value

def correlation(data, threshold):
    col_corr=set()
    corr_matrix = data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i,j])>threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
X = df.drop('Target', axis=1)
Y = df['Target']

#Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X,Y, train_size=0.7, random_state=11, shuffle=True)
```

```python
#setting up threshold as 70%  

corr_features = correlation(X_train,0.7)
len(set(corr_features))
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from scipy import stats


X2 = sm.add_constant(X_train)
est = sm.OLS(y_train, X2)
est2 = est.fit()
print(est2.summary())
```

```python
columns = ["F2","F4", "F5", "F6", "F7", "F8", "F9", "F10", "F12", "F13", "F16", "F17", "F18", "F22", "F24", "F26", "F28", "F29", "F31", "F32", "F33", "F34", "F35", "F36"]
X_train = X_train.drop(columns, axis=1)
X_test = X_test.drop(columns, axis=1)
```

P-value that is <= 0.05 is significant.

In the above regression model, the features with p-value more than 0.05 were removed, since they are not significant.

```python
X_train.shape
```

```python
# Standardisation

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression(normalize=True)
linear_regressor.fit(X_train, y_train)

y_predict = linear_regressor.predict(X_test)
lr_acc = r2_score(y_test, y_predict)
r2_score(y_test, y_predict)
```

Linear Regression score is 54%

# Different Algorithms:

```python
#SVR

from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

svr_regressor = SVR(kernel='linear', C=5, gamma='auto')

#Choosing best Parm for SVR using GridSearchCV

parameters = {'kernel':('linear', 'rbf','poly','sigmoid'), 'C':[1,3,5,10,15], 'gamma':('auto','scale') }
clf = GridSearchCV(svr_regressor, parameters)
clf.fit(X_train, y_train)
```

```python
#Best Parameters
clf.best_params_
```

```python
#SVR with best parameter

svr_regressor = SVR(kernel='linear', C=10, gamma='auto')
svr_regressor.fit(X_train, y_train)

y_predict1 = svr_regressor.predict(X_test)
svr_acc = r2_score(y_test, y_predict1)
r2_score(y_test, y_predict1)
```

```python
#Gradient Boosting Regressor

from sklearn.ensemble import GradientBoostingRegressor

GBR = GradientBoostingRegressor()
GBR.fit(X_train, y_train)

y_predict2 = GBR.predict(X_test)
gbr_acc = r2_score(y_test, y_predict2)
r2_score(y_test, y_predict2)
```

```python
#Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor

forest_regressor = RandomForestRegressor(n_estimators = 300, random_state = 0)
forest_regressor.fit(X_train, y_train)

y_predict3 = forest_regressor.predict(X_test)
rfr_acc = r2_score(y_test, y_predict3)
r2_score(y_test, y_predict3)
```

```python
#LGBM Regressor

from lightgbm import LGBMRegressor

LGBM = LGBMRegressor()
LGBM.fit(X_train, y_train)

y_predict4 = LGBM.predict(X_test)
lgbm_acc = r2_score(y_test, y_predict4)
r2_score(y_test, y_predict4)
```

```python
#XGBoost Regressor

from xgboost.sklearn import XGBRegressor

XGBR = XGBRegressor()
XGBR.fit(X_train, y_train)

y_predict5 = XGBR.predict(X_test)
xgb_acc = r2_score(y_test, y_predict5)
r2_score(y_test, y_predict5)
```

# Choosing the best model 

Based on the accuracy from different model. Let us plot, compare and choose the best model.

```python
models = ["Linear Regression","SV R", "GB R", "RF R", "LGBM R", "XGB R"]
accuracies = [lr_acc, svr_acc, gbr_acc, rfr_acc, lgbm_acc, xgb_acc]
plt.figure(figsize=(12,5))
plt.bar(models,accuracies)
plt.xlabel("Models")
plt.ylabel("Accuracy")
```

On Comparison, we could find that LGBM Regressor is having a highest accuracy(85%) of all.
